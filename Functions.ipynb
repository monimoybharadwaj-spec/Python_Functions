```python
# Question 1. What is the difference between a function and a method in Python?

# Answer: A function is a reusable block of code that performs a task and can be used anywhere in the program. Example:

def add(a, b):
    return a + b

print(add(5, 3))

# A method is a function that is used with a specific value or data type, like a string, list, or number. Example:

name = "Python"
print(name.upper())
```

```python
# Question 2. Explain the concept of function arguments and parameters in Python.

# Answer: A parameter is a variable in the function definition that accepts a value when the function is called. Example:

def greet(name):
    print("Hello, " + name)
# Here, 'name' is a parameter.

# An argument is the actual value passed to the function when it is called. Example:

greet("Alice")
# Here, "Alice" is the argument.
```

```python
# Question 3. What are the different ways to define and call a function in Python?

# Answer: In Python, functions can be defined and called in several ways:

# (1) Functions without parameters:
def greet():
    print("Hello, World!")

greet()

# (2) Functions with parameters:
def greet(name):
    print("Hello, " + name)

greet("Max")

# (3) Functions with default arguments:
def sum1(b, c, a = 0):
    return a+b+c

print(sum1(3,5,5))

# (4) Functions with return values:
def func():
    calc = 2+3+6*5
    return calc

print(func())

# (5) Calling functions with keyword arguments:
def info(name, age):
    print(f"Name: {name}, Age: {age}")

info(age=20, name="Max")
```

```python
# Question 4. What is the purpose of the `return` statement in a Python function?

# Answer: The 'return' statement is used to send a value back to the part of the program that called the function. Example:

def square(x):
    return x * x  # Returns the square of x

print(square(4))
```

```python
# Question 5. What are iterators in Python and how do they differ from iterables?

# Answer: An iterable is any Python object such as a list, tuple, or string that can return its items one at a time using loops, while
# an iterator is an object that goes through the items of an iterable one by one.

# Example of iterable : List is an iterable object

lis = [1,2,"Miku"]
for i in lis:
    print(i)

# Example of iterator :

s = "MikuTheGreat"
for i in s:
    print(i)
```

```python
# Question 6. Explain the concept of generators in Python and how they are defined.

# Answer: A generator is a function that returns values one at a time using yield, instead of returning all values at once. Example:

def square_gen(n):
    for i in range(n):
        yield i**2

gen = square_gen(4)
print(list(gen))
```

```python
# Question 7. What are the advantages of using generators over regular functions?

# Answer: Generators are memory-efficient, compute values only when needed, can handle infinite sequences, improve performance
# for large data, and make code simpler compared to regular functions. Example:

# Regular function for squaring the elements of a list
def square_no(n):
    result=[]
    for i in range(n):
        result.append(i**2)
    return result

print(square_no(4))

# Using generators for squaring the elements of a list
def square_gen(n):
    for i in range(n):
        yield i**2

gen = square_gen(4)
print(list(gen))
```

```python
# Question 8. What is a lambda function in Python and when is it typically used?

# Answer: A lambda function is a small anonymous function defined with lambda, typically used for short, simple operations or with functions like map() and filter(). Example:

sqr = lambda x: x**2
print(sqr(3))
```

```python
# Question 9. Explain the purpose and usage of the `map()` function in Python.

# Answer: The map() function is used to apply a function to every item of an iterable (like a list, tuple, or string) and return a new iterator with the results.
# It is useful for performing the same operation on all elements without using a loop explicitly. Example:

def add(x):
    return x + 10

l = [1, 2, 3, 4, 5]
map(add,l)
list(map(add,l))
```

```python
# Question 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

# Answer: In Python, map() applies a function to all items. Example:

print(list(map(lambda x: x*2, [1,2,3])))

# filter() selects items based on a condition. Example:

print(list(filter(lambda x: x%2==0, [1,2,3,4])))

# reduce() combines items into a single result. It will always take two arguments. It requires "from functools import reduce". Example:

from functools import reduce
reduce(lambda x,y: x+y, [1,2,3,4])
```

```python
# Question 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

# Answer:
def sum_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

lst1 = [1, 2, 3, 10, 7, 6]
print("Sum of even numbers:", sum_even_numbers(lst1))
```

```python
# Question 2. Create a Python function that accepts a string and returns the reverse of that string.

# Answer:
def reverse_string(str1):
    return str1[::-1]

str1 = "Hello, World!"
print("Reversed string:", reverse_string(str1))
```

```python
# Question 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

# Answer:
def sq_num(numbers):
    return list(map(lambda x: x**2, numbers))

lst2 = [1, 2, 3, 4, 5, 6]
print("Squared List:", sq_num(lst2))
```

```python
# Question 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

# Answer:
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

for n in range(1, 201):
    print(f"{n} is prime: {is_prime(n)}")
```

```python
# Question 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

# Answer:
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = FibonacciIterator(20)
for num in fib:
    print(num, end=" ")
```

```python
# Question 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

# Answer:
def powers_of_two(n):
  for i in range(0,n+1):
    yield 2**i

for power in powers_of_two(5):
  print(power)
```

```python
# Question 7.  Implement a generator function that reads a file line by line and yields each line as a string.

# Answer:

def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

# Example usage
for line in read_file_lines("example.txt"):
    print(line)
```

```python
# Question 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Answer:
lst3 = [(1, 5), (3, 1), (4, 7), (2, 3)]

sorted_lst3 = sorted(lst3, key=lambda x: x[1])

print("Sorted list:", sorted_lst3)
```

```python
# Question 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# Answer:
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

lst4 = [26, 28, 30, 32, 34, 36]
fahrenheit = list(map(celsius_to_fahrenheit,lst4))
print("Temperature in Fahrenheit:", fahrenheit)
```

```python
# Question 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Answer:
text = "It is a wonderful evening."
vowels = "aeiouAEIOU"

result = ""
for ch in filter(lambda c: c not in vowels, text):
    result += ch

print("String without vowels:", result)
```
